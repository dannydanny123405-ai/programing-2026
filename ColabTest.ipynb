## 잠실 야구장 좌석 추천 질문 및 카테고리

사용자에게 물어볼 질문과 그에 따라 추천될 좌석 카테고리를 정리해 봅시다.

### 질문 목록:
1.  **테이블이 있는 좌석을 원하시나요?** (예/아니오)
    *   `예` 선택 시: 프리미엄석, 테이블석 고려
    *   `아니오` 선택 시: 일반석, 응원석 등 다른 좌석 고려
2.  **파울볼 또는 홈런볼을 잡고 싶으신가요?** (예/아니오)
    *   `예` 선택 시: 외야석, 익사이팅존 등 타구 위험이 있는 좌석 고려
3.  **응원가를 같이 부르며 시끄러운 분위기를 즐기고 싶으신가요?** (예/아니오)
    *   `예` 선택 시: 응원단상 근처, 레드석 등 응원 열기가 뜨거운 좌석 고려
4.  **원정팀 팬이신가요?** (예/아니오)
    *   `예` 선택 시: 3루 응원석, 원정팀 지정 좌석 고려
5.  **한눈에 그라운드 전체를 보고 싶으신가요?** (예/아니오)
    *   `예` 선택 시: 중앙 지정석, 블루석 상단 등 시야가 좋은 좌석 고려

### 주요 좌석 카테고리 (예시):
*   **프리미엄석/테이블석**: 편안한 관람과 테이블 이용. (1번 질문 관련)
*   **익사이팅존/외야석**: 파울볼, 홈런볼 기회. (2번 질문 관련)
*   **응원단상석/레드석**: 열정적인 응원 분위기. (3번 질문 관련)
*   **원정팀 응원석**: 원정팀 팬들을 위한 공간. (4번 질문 관련)
*   **중앙 지정석/블루석**: 넓은 시야로 전체 경기 조망. (5번 질문 관련)
*   **블루석/네이비석**: 가성비 좋은 일반 좌석.

```python
import gradio as gr

def recommend_seat(
    has_table: str,
    want_foul_ball: str,
    want_cheer: str,
    is_away_fan: str,
    want_full_view: str
) -> str:
    """
    잠실 야구장 좌석을 추천하는 함수.

    Args:
        has_table (str): 테이블 있는 좌석을 원하는지 (예/아니오)
        want_foul_ball (str): 파울볼/홈런볼을 잡고 싶은지 (예/아니오)
        want_cheer (str): 응원가를 부르며 시끄러운 분위기를 원하는지 (예/아니오)
        is_away_fan (str): 원정팀 팬인지 (예/아니오)
        want_full_view (str): 그라운드 전체를 한눈에 보고 싶은지 (예/아니오)

    Returns:
        str: 추천 좌석
    """

    recommendations = []

    # 1. 테이블 유무
    if has_table == '예':
        recommendations.append('프리미엄석 또는 테이블석')
    else:
        # 테이블이 필요 없다면 다른 좌석 고려
        pass

    # 2. 파울볼/홈런볼
    if want_foul_ball == '예':
        recommendations.append('익사이팅존 또는 외야석')

    # 3. 응원 분위기
    if want_cheer == '예':
        recommendations.append('응원단상 근처 좌석 (블루석, 레드석 일부)')

    # 4. 원정팀 팬
    if is_away_fan == '예':
        recommendations.append('원정 응원석 (3루 레드석/오렌지석)')

    # 5. 전체 그라운드 시야
    if want_full_view == '예':
        recommendations.append('중앙 지정석 (블루석 중앙 구역) 또는 스카이박스')

    if not recommendations:
        return "가성비 좋은 일반 좌석 (블루석, 네이비석)을 추천합니다."
    else:
        return "다음 좌석들을 고려해보세요: " + ", ".join(list(set(recommendations))) + "\n\n(참고: 정확한 좌석 배치는 구장 정책 및 예매 상황에 따라 달라질 수 있습니다.)"

# Gradio 인터페이스 생성
iface = gr.Interface(
    fn=recommend_seat,
    inputs=[
        gr.Radio(['예', '아니오'], label='1. 테이블이 있는 좌석을 원하시나요?'),
        gr.Radio(['예', '아니오'], label='2. 파울볼 또는 홈런볼을 잡고 싶으신가요?'),
        gr.Radio(['예', '아니오'], label='3. 응원가를 같이 부르며 시끄러운 분위기를 즐기고 싶으신가요?'),
        gr.Radio(['예', '아니오'], label='4. 원정팀 팬이신가요?'),
        gr.Radio(['예', '아니오'], label='5. 한눈에 그라운드 전체를 보고 싶으신가요?')
    ],
    outputs="text",
    title="잠실 야구장 좌석 추천 앱",
    description="몇 가지 질문에 답하여 잠실 야구장 최고의 좌석을 추천받으세요!"
)

iface.launch(debug=True, share=True)
```

```python
import gradio as gr

def recommend_seat(
    has_table: str,
    want_foul_ball: str = '아니오',
    want_cheer: str = '아니오',
    is_away_fan: str = '아니오',
    want_full_view: str = '아니오'
) -> str:
    # Define some dummy price ranges (예시 가격, 실제와 다를 수 있음)
    prices = {
        '프리미엄석': '50,000원 ~ 80,000원',
        '테이블석': '40,000원 ~ 70,000원 (1인 기준)',
        '익사이팅존': '35,000원 ~ 50,000원',
        '외야석': '15,000원 ~ 25,000원',
        '응원단상 근처 좌석': '25,000원 ~ 40,000원', # 블루석, 레드석
        '원정 응원석': '20,000원 ~ 35,000원', # 3루 레드석/오렌지석
        '중앙 지정석': '30,000원 ~ 50,000원', # 블루석 중앙 구역
        '스카이박스': '고가 (단체 예약)',
        '블루석/네이비석': '20,000원 ~ 35,000원'
    }

    recommendations = set()

    # 1. 테이블 유무에 따른 1차 판단
    if has_table == '예':
        # 테이블이 있는 좌석을 원하면, 즉시 추천하고 종료
        premium_price = prices.get('프리미엄석', '가격 정보 없음')
        table_price = prices.get('테이블석', '가격 정보 없음')
        return f"테이블이 있는 좌석을 원하시면 다음을 추천합니다:\n\n" \
               f"- 프리미엄석 (예상 가격: {premium_price})\n" \
               f"- 테이블석 (예상 가격: {table_price})\n\n" \
               f"(참고: 정확한 좌석 배치는 구장 정책 및 예매 상황에 따라 달라질 수 있습니다.)"

    # '테이블 없음'일 경우에만 나머지 질문들을 고려
    if want_foul_ball == '예':
        recommendations.add(f"익사이팅존 (예상 가격: {prices.get('익사이팅존', '가격 정보 없음')})")
        recommendations.add(f"외야석 (예상 가격: {prices.get('외야석', '가격 정보 없음')})")

    if want_cheer == '예':
        recommendations.add(f"응원단상 근처 좌석 (블루석, 레드석 일부) (예상 가격: {prices.get('응원단상 근처 좌석', '가격 정보 없음')})")

    if is_away_fan == '예':
        recommendations.add(f"원정 응원석 (3루 레드석/오렌지석) (예상 가격: {prices.get('원정 응원석', '가격 정보 없음')})")

    if want_full_view == '예':
        recommendations.add(f"중앙 지정석 (블루석 중앙 구역) (예상 가격: {prices.get('중앙 지정석', '가격 정보 없음')})")
        recommendations.add(f"스카이박스 (예상 가격: {prices.get('스카이박스', '가격 정보 없음')})")

    if not recommendations:
        return f"선택하신 조건에 해당하는 특별한 좌석은 없으나, 가성비 좋은 일반 좌석 (블루석, 네이비석)을 추천합니다. (예상 가격: {prices.get('블루석/네이비석', '가격 정보 없음')})"
    else:
        return "다음 좌석들을 고려해보세요:\n\n" + "\n".join([f"- {rec}" for rec in list(recommendations)]) + \
               "\n\n(참고: 정확한 좌석 배치는 구장 정책 및 예매 상황에 따라 달라질 수 있습니다.)"

# Gradio Blocks를 사용하여 조건부 UI 구현
with gr.Blocks(title="잠실 야구장 좌석 추천 앱") as demo:
    gr.Markdown("<h1 style='text-align: center; color: #333;'>⚾ 잠실 야구장 좌석 추천 앱 ⚾</h1>")
    gr.Markdown("<p style='text-align: center; color: #555;'>몇 가지 질문에 답하여 잠실 야구장 최고의 좌석을 추천받으세요!</p>")

    with gr.Row():
        q1_input = gr.Radio(['예', '아니오'], label='1. 테이블이 있는 좌석을 원하시나요?', value=None, interactive=True)

    # 나머지 질문들과 '좌석 추천 받기' 버튼을 포함하는 그룹 (초기에는 숨김)
    with gr.Column(visible=False) as other_elements_group:
        gr.Markdown("---<br/>")
        gr.Markdown("### 추가 질문")
        with gr.Row():
            q2_input = gr.Radio(['예', '아니오'], label='2. 파울볼 또는 홈런볼을 잡고 싶으신가요?', value=None, interactive=True)
        with gr.Row():
            q3_input = gr.Radio(['예', '아니오'], label='3. 응원가를 같이 부르며 시끄러운 분위기를 즐기고 싶으신가요?', value=None, interactive=True)
        with gr.Row():
            q4_input = gr.Radio(['예', '아니오'], label='4. 원정팀 팬이신가요?', value=None, interactive=True)
        with gr.Row():
            q5_input = gr.Radio(['예', '아니오'], label='5. 한눈에 그라운드 전체를 보고 싶으신가요?', value=None, interactive=True)

        submit_btn = gr.Button("좌석 추천 받기")

    # 추천 좌석 출력 텍스트 상자 (더 크게 설정)
    output_text = gr.Textbox(label="🌟 추천 좌석", lines=8, interactive=False, container=True)

    # q1_input 값이 변경될 때 UI 업데이트 및 초기화 함수
    def update_ui_and_recommend_if_table(q1_val):
        # Reset all other question inputs to None and output_text to empty
        reset_q_updates = [gr.update(value=None) for _ in range(4)] # q2, q3, q4, q5

        if q1_val == '예':
            # Hide other questions group
            # Immediately call recommend_seat for table seats and update output
            table_recommendation = recommend_seat(q1_val) # Only pass q1_val
            return (gr.update(visible=False),) + tuple(reset_q_updates) + (gr.update(value=table_recommendation),)
        elif q1_val == '아니오':
            # Show other questions group
            # Clear output_text
            return (gr.update(visible=True),) + tuple(reset_q_updates) + (gr.update(value=""),)
        else: # q1_input is None (initial state)
            return (gr.update(visible=False),) + tuple(reset_q_updates) + (gr.update(value=""),)


    # q1_input의 변경 이벤트에 UI 업데이트 함수 연결
    q1_input.change(
        update_ui_and_recommend_if_table,
        inputs=[q1_input],
        outputs=[other_elements_group, q2_input, q3_input, q4_input, q5_input, output_text]
    )

    # '좌석 추천 받기' 버튼 클릭 이벤트에 추천 로직 연결 (q1_input이 '아니오'일 경우에만 유효)
    submit_btn.click(
        recommend_seat,
        inputs=[q1_input, q2_input, q3_input, q4_input, q5_input],
        outputs=output_text
    )

demo.launch(debug=True, share=True)
```
